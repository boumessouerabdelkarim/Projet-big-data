```python
data_frame.createOrReplaceTempView("cardio_table")
resultat_sql = spark.sql("SELECT * FROM cardio_table ")
resultat_sql.toPandas()
```

```python

pandas_df=resultat_sql.toPandas()
sns.countplot(data=pandas_df, x="cardio")
plt.title("Distribution of heart attack")
plt.show()
```

```python

filtered_df = resultat_sql.filter(data_frame.cardio == "YES")

# Regrouper par genre et calculer le nombre d'occurrences
grouped_df = filtered_df.groupBy("gender").count()

# Convertir en DataFrame Pandas pour la visualisation
pandas_df = grouped_df.toPandas()

# Calculer le pourcentage total
total = pandas_df['count'].sum()

# Ajouter une colonne pour le pourcentage
pandas_df['percentage'] = (pandas_df['count'] / total) * 100

# Création du camembert avec le nombre d'occurrences
def autopct(pct):
    """Fonction personnalisée pour afficher le pourcentage et le nombre d'occurrences."""
    value = round(pct * total / 100)
    return f"{pct:.1f}% ({value})"

plt.pie(pandas_df['percentage'], labels=pandas_df['gender'], autopct=autopct)
plt.title("Pourcentage de cardio-patients par genre")
plt.show()
```

```python
def convert_days_to_years(age_in_days):
    return floor(age_in_days / 365.25).cast("int")

# Application de la fonction et remplacement direct de la colonne "age"
data_frame = data_frame.withColumn("age", convert_days_to_years(col("age")))

# Affichage du DataFrame modifié
data_frame.toPandas()
```

```python
data_frame = data_frame.withColumn("Categories_age", 
                                  when(col("age")<14, "Enfants")
                                  .when((col("age")>14) & (col("age")<30), "Adolescents")
                                  .when((col("age")>30) & (col("age")<60), "Adultes")
                                  .otherwise("Aînés")
                                 )

# Affichage du DataFrame modifié
data_frame.toPandas()
```

```python

resultat_with_cardio=data_frame.filter(col("cardio")=="YES")
total_cardio=resultat_with_cardio.count()
resultat=resultat_with_cardio.groupBy("gender","Categories_age").count()
rrd1=resultat.withColumn("pourcentage",(col("count")/total_cardio)*100)
rrd1.show()
```

```python


rrd1.toPandas().pivot_table(index='Categories_age', columns='gender', values='pourcentage').plot(kind='bar', stacked=True)
plt.title("Répartition des cas par genre et catégorie d'âge")
plt.xlabel("Catégorie d'âge")
plt.ylabel("Pourcentage")
plt.show()
```

```python
resultat_smoke=resultat_with_cardio.groupBy("gender","Categories_age","smoke").count()
rrd2=resultat_smoke.withColumn("pourcentage",(col("count")/total_cardio)*100)
rrd2.show()
```

```python
rrd2.toPandas().pivot_table(index='Categories_age', columns=['gender', 'smoke'], values='pourcentage').plot(kind='bar', stacked=True)
plt.title("Répartition du tabagisme par genre et catégorie d'âge")
plt.xlabel("Catégorie d'âge")
plt.ylabel("Pourcentage")
plt.show()
```

```python
def remplacer_genre_binaire(dataframe, colonne_genre):

  return dataframe.withColumn(colonne_genre, 
                              when(col(colonne_genre) == "YES", 1)
                              .when(col(colonne_genre) == "NO", 0)
                              .otherwise(col(colonne_genre))
                             )

data_frame=remplacer_genre_binaire(data_frame, "smoke")
data_frame=remplacer_genre_binaire(data_frame, "alco")
data_frame=remplacer_genre_binaire(data_frame, "active")
data_frame=remplacer_genre_binaire(data_frame, "cardio")

# Affichage du DataFrame modifié
data_frame.toPandas()
```

```python
def remplacer_genre(dataframe, colonne_genre):

  return dataframe.withColumn(colonne_genre, 
                              when(col(colonne_genre) == "normal",1 )
                              .when(col(colonne_genre) == "above normal",2 )
                               .when(col(colonne_genre) == "well above normal",3 )
                              .otherwise(col(colonne_genre))
                             )

data_frame=remplacer_genre(data_frame, "cholesterol")
data_frame=remplacer_genre(data_frame, "gluc")

# Affichage du DataFrame modifié
data_frame.toPandas()
```
